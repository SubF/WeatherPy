```python
# Import the dependencies.
import pandas as pd
import numpy as np
from citipy import citipy
import requests
import time
from datetime import datetime

from config import weatherkey
```

```python
# Query URL
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weatherkey
```

```python
# Create a set of random latitude and longitude combinations.
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
lngs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lats_lngs = zip(lats, lngs)

#Create list of lat and lngs
coordinates = list(lats_lngs)
```

```python
#Create list of cities
cities = []

#Check for nearest cities with citipy
for coordinates in coordinates:
    city = citipy.nearest_city(coordinates[0], coordinates[1]).city_name
    
    #Add unique cities to list
    if city not in cities:
        cities.append(city)
        
len(cities)
```
